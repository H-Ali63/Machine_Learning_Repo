# INTRODUCTION TO STATISTICS

Statistics is the science of **collecting, organizing, analyzing, interpreting, and presenting** data to make decisions.

### Types of Statistics

Descriptive Statistics – Summarizes and describes data

Inferential Statistics – Draws conclusions about a population using sample data

### 1. DESCRIPTIVE STATISTICS

Descriptive statistics help us **understand data at a glance.**

### a. Measure of Central Tendency

Shows the center or typical value of a dataset.

            Mean=∑x/n

Example:
Data: 2, 4, 6, 8
Mean = (2+4+6+8)/4 = 5

Sensitive to outliers

Used for numerical data

### b. Median

The middle value when data is arranged in ascending order.


Example:
Data: 3, 5, 7
Median = 5

For even values:
Data: 3, 5, 7, 9
Median = (5+7)/2 = 6

Not affected by outliers

Useful for skewed data

### c. Mode

The most frequently occurring value.

Example:
Data: 1, 2, 2, 3, 4
Mode = 2


Used for categorical data

Can be more than one mode

### Comparison

| Measure | Best Used When          |
| ------- | ----------------------- |
| Mean    | Data is symmetric       |
| Median  | Data is skewed          |
| Mode    | Categories or frequency |


### 2. Measure of Dispersion

Shows how spread out the data is.

### a) Range
Range = Max − Min

Example:

Data: 5, 10, 15


Range = 15 − 5 = 10


### b) Variance

Measures average squared deviation from the mean.



Where:
𝜎^2= Population Variance
𝑥= Each data value
μ = Population mean
𝑁= Total number of observations

Given population data: 2, 4, 6, 8

Step 1: Find the mean (𝜇)

Step 2: Find deviations from mean
| x | (x - μ)    | ((x - μ)^2)   |
| - | ---------- | ------------- |
| 2 | 2 − 5 = −3 | 9             |
| 4 | 4 − 5 = −1 | 1             |
| 6 | 6 − 5 = 1  | 1             |
| 8 | 8 − 5 = 3  | 9             |

∑(x−μ)^2 = 9 + 1 + 1 + 9 = 20


Step 3: Apply population variance formula 𝜎^2 = 20 / 4 = 5



### c) Standard Deviation

Square root of variance.

            σ = √σ**2


Most commonly used dispersion measure

Lower SD → data close to mean



### d) Interquartile Range (IQR)

    IQR = 𝑄3 − 𝑄1


Used to detect outliers

### 3. Measure of Shape

Describes data distribution shape.



### a) Skewness

    Positive Skew → Tail on right

    Negative Skew → Tail on left

    Zero Skew → Symmetric

| Distribution  | Relation             |
| ------------- | -------------------- |
| Positive Skew | Mean > Median > Mode |
| Negative Skew | Mean < Median < Mode |

### b) Kurtosis

Measures peakedness of distribution.

    Leptokurtic – Sharp peak

    Mesokurtic – Normal

    Platykurtic – Flat

## SAMPLING TECHNIQUES

Sampling = Selecting a subset from a population.

Probability Sampling

    Simple Random Sampling → Samples are chosen randomly, and every member has an equal chance of selection.
    Stratified Sampling → Population is divided into subgroups (strata), and samples are taken from each group.
    Systematic Sampling → Samples are selected at a fixed regular interval (every kth element) after a random start.
    Cluster Sampling → Population is divided into clusters, and entire clusters are randomly selected.

Non-Probability Sampling

    Convenience Sampling -> Easy access
    Judgment Sampling    -> Expert choice
    Snowball Sampling    -> Referral based
    Quota Sampling       -> Fixed proportions

# PROBABILITY DISTRIBUTIONS
# 7. Discrete Probability Distribution
Used when outcomes are countable.



(Used when outcomes are **countable**)


### Bernoulli Distribution

Single trial with **two outcomes** (success/failure).

**Example:**

* Tossing a coin once (Head or Tail)
* Pass / Fail result


### Binomial Distribution

Fixed number of independent trials with **two outcomes**.

**Example:**

* Tossing a coin 10 times and counting heads
* Number of defective items in a batch


### Poisson Distribution

Used for **rare events** in a fixed interval.

**Example:**

* Number of calls received in a call center per hour
* Number of accidents at a junction per day


### Discrete Uniform Distribution

All outcomes have **equal probability**.

**Example:**

* Rolling a fair die
* Selecting a random card from 1 to 10


# 8 Continuous Probability Distributions

(Used when outcomes are **measurable**)


### Normal Distribution

Bell-shaped and symmetric.

**Example:**

* Heights of students
* IQ scores
* Measurement errors


### Uniform Distribution (Continuous)

All values within a range are equally likely.

**Example:**

* Random number between 0 and 1
* Waiting time between 0–10 minutes


### Exponential Distribution

Models **time between events**.

**Example:**

* Time between customer arrivals
* Lifetime of a light bulb


### Gamma Distribution

Used for **waiting time until multiple events occur**.

**Example:**

* Time until 3 customers arrive
* Total rainfall in a month



# HYPOTHESIS TESTING

### 9. Hypothesis

A claim about population parameters.

    a) Null Hypothesis (H₀)
        No effect
        No difference
        Default assumption

    b) Alternative Hypothesis (H₁)
        Opposite of H₀
        What we want to prove



Type I and Type II Errors


| Error Type  | Meaning                    |
| ----------- | -------------------------- |
| Type I (α)  | Reject H₀ when it is true  |
| Type II (β) | Accept H₀ when it is false |

### 10. P-Value

Probability of observing results as extreme as current results, assuming H₀ is true.

    Small p-value → strong evidence against H₀

    Large p-value → weak evidence

### 11. Significance Level (α)

    Threshold for rejecting H₀
    Common values: 0.05, 0.01

 **Decision Rule**

If p ≤ α → Reject H₀
If p > α → Fail to reject H₀
